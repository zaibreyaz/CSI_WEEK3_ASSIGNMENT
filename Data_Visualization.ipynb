# Data Visualization using Iris Dataset

## Importing libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Importing Dataset

```python
dataset = sns.load_dataset('iris')
dataset.head()
```

#### Visualization of relationship between features

```python
sns.pairplot(dataset, hue='species')
plt.show()
```

#### Visualizing distribution of features by species

- Sepal Length

```python

plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
sns.boxplot(data=dataset, x='species', y='sepal_length')
plt.title('BoxPlot of Distribution of Sepal Length by Species')

plt.subplot(2, 2, 2)
sns.violinplot(data=dataset, x='species', y='sepal_length')
plt.title('ViolinPlot of Distribution of Sepal Length by Species')
plt.show()
```

- Sepal Width

```python
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
sns.boxplot(data=dataset, x='species', y='sepal_width')
plt.title('BoxPlot of Distribution of Sepal Width by Species')

plt.subplot(2, 2, 2)
sns.violinplot(data=dataset, x='species', y='sepal_width')
plt.title('ViolinPlot of Distribution of Sepal width by Species')
plt.show()
```

- Petal Length

```python
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
sns.boxplot(data=dataset, x='species', y='petal_length')
plt.title('BoxPlot of Distribution of Petal Length by Species')

plt.subplot(2, 2, 2)
sns.violinplot(data=dataset, x='species', y='petal_length')
plt.title('ViolinPlot of Distribution of Petal Length by Species')
plt.show()
```

- Petal Width

```python
plt.figure(figsize=(12, 10))
plt.subplot(2, 2, 1)
sns.boxplot(data=dataset, x='species', y='petal_width')
plt.title('BoxPlot of Distribution of Petal Width by Species')

plt.subplot(2, 2, 2)
sns.violinplot(data=dataset, x='species', y='petal_width')
plt.title('ViolinPlot of Distribution of Petal Width by Species')
plt.show()
```

#### Distribution of Features by Species

```python
plt.figure(figsize=(12, 8))
plt.suptitle('Distribution of Features by Species')

plt.subplot(2, 2, 1)
sns.histplot(dataset, x='sepal_length', hue='species', kde=True, bins=15)

plt.subplot(2, 2, 2)
sns.histplot(dataset, x='sepal_width', hue='species', kde=True, bins=15)

plt.subplot(2, 2, 3)
sns.histplot(dataset, x='petal_length', hue='species', kde=True, bins=15)

plt.subplot(2, 2, 4)
sns.histplot(dataset, x='petal_width', hue='species', kde=True, bins=15)

plt.show()
```

#### Heatmap

```python
plt.figure(figsize=(10, 6))

sns.heatmap(dataset.corr())
plt.title('Correlation Between Features')
plt.show()
```
